```python
import numpy as np
import pandas as pd
```

```python
data_one = {'A':['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']}
```

```python
data_two = {'C' : ['C0', 'C1', 'C2', 'C3'], 'D':['D0', 'D1', 'D2', 'D3']}
```

```python
one = pd.DataFrame(data_one)
```

```python
two = pd.DataFrame(data_two)
```

```python
one
```

```python
two
```

```python
#Concatenating columns
```

```python
pd.concat([one,two], axis=1)
```

```python
#Concatinating rows
```

```python
two.columns = one.columns
```

```python
two
```

```python
mydf = pd.concat([one,two],axis=0)
```

```python
data_a = {'Reg_id': ['1', '2', '3', '4'], 'name': ['Andrew', 'Bob', 'Charlie', 'David']}
```

```python
data_b = {'log_id': ['1', '2', '3', '4'], 'name': ['Xavier', 'Andrew', 'Yolanda', 'Bob']}
```

```python
registration = pd.DataFrame(data_a)
```

```python
logins = pd.DataFrame(data_b)
```

```python
registration
```

```python
logins
```

```python
#inner merge
```

```python
pd.merge(registration,logins,how='inner',on='name')
```

```python
  #left merge
```

```python
pd.merge(registration,logins,how='left',on='name')
```

```python
pd.merge(registration,logins,how='right',on='name')
```

```python
pd.merge(registration,logins,how='outer',on='name')
```
