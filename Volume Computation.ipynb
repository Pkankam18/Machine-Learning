```python
#NB:T_L - Thickness left
    #T_R - Thickness Right
    #T_C - Thickness Center
    #Mean_T - Mean Thickness 
```

```python
df = pd.read_csv('C:\\Users\\Nana Qwasi\\Desktop\\Tutorials\\Programming\\Projects\\Volume data.csv')
```

```python
df.head()
```

```python
df['T_L'] = df['Fin.lev.L'] - df['I_L_Left']
```

```python
df.head()
```

```python
df['T_C'] = df['Fin.lev.C'] - df['I_L_Centre']
```

```python
df['T_R'] = df['Fin.lev.R'] - df['I_L_Right']
```

```python
df.head()
```

```python
df['Average'] = (df['T_L'] + df['T_C'] + df['T_R']) / 3
```

```python
df.head()
```

```python
df['Avearge'] = df['Average'].round(3)
```

```python
df.head()
```

```python
df['Mean_T'] = (df['Average'] + df['Average'].shift(1)) / 2
```

```python
df.head()
```

```python
df['Mean_T'] = df['Mean_T'].round(3)
```

```python
df.loc[3,'I_L_Right'] = 197.081
```

```python
df.head()
```

```python
df.to_csv('C:\\Users\\Nana Qwasi\\Desktop\\Tutorials\\Programming\\Projects\\Volume data.csv',index=False)
```

```python
df['T_R'] = df['Fin.lev.R'] - df['I_L_Right']
```

```python
df['Mean_W'] = (df['Width'] + df['Width'].shift(1)) / 2
```

```python
df['Volume'] = (df['Chainage'] - df['Chainage'].shift(1)) * df['Mean_T'] * df['Mean_W']
```
